Step 5: Creating NumPy Arrays
One-dimensional and two-dimensional arrays are created using NumPy.

```python
import numpy as np
arr1 = np.array([10,20,30,40,50])
print("1D array:",arr1)
arr2=np.array([[1,2,3],[4,5,6]])
print("2D array:\n",arr2)



```

step 6:Array properties and indexing and accessing elements


Array properties are used to understand the structure of the array.
Shape – gives dimensions
Size – total number of elements
Data type – type of elements

```python
print("first element of 1D array:",arr1[0])
print("elements from index 1 to 3:",arr1[1:4])
print("element at row 1,col 2:",arr2[0,1])

```

step 7:Mathematical and statical operations

- np.sum() → sum of array elements
- np.mean() → mean of array elements
- np.max() → maximum element
- np.min() → minimum element
- axis parameter:
    - axis=0 → sum along columns
    - axis=1 → sum along rows

```python
print("sum of 1D array:",np.sum(arr1))
print("mean of 1D array:",np.mean(arr1))
print("maximum value in 1D array:",np.max(arr1))
print("minimum value in 1D array:",np.min(arr1))


```

```python

print("sum along rows OF 2D array:",np.sum(arr2,axis=1))
print("sum along columns of 2D array:",np.sum(arr2,axis=0))

```

Step 9: Broadcasting:

- Broadcasting allows arithmetic operations between arrays of different shapes.
- Smaller array is broadcasted to match the shape of the larger array.

```python
arr4 = np.array([1,2,3])
broadcasted = arr4 + 5
print("broadcasted array:", broadcasted)
```

Step 10: Saving and Loading Arrays


- NumPy arrays can be saved to a file using np.save().
- np.load() is used to load saved arrays.

```python
np.save("sample.npy", arr1)
loaded = np.load("sample.npy")
print(loaded)
```

Step 11: Element-wise Operations

- NumPy supports element-wise operations like addition, multiplication, and exponentiation.
- Operations are faster than Python lists.

```python

print(arr1 ** 2)


print(arr1 + arr1)
```

Step 12: Performance Comparison with Python Lists


- NumPy operations are faster than Python lists for large data.
- Demonstrates time difference for incrementing elements.

```python
import time

# Python list
py_list = list(range(1000))
start = time.time()
py_list = [x+1 for x in py_list]
print("Python list time:", round(time.time()-start,6))

# NumPy array
np_array = np.array(range(100000))
start = time.time()
np_array = np_array + 1
print("NumPy array time:", round(time.time()-start,6))
```
